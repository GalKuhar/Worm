## Povezava med številom besed in tem ali je fic zaključen:

Razdelili bomo fice v skupine po besedah. Ker je razpon zelo velik, bomo to naredili logaritmično:

### Povprečna dolžina

```python
fics['besede'].mean()
```

```python
def naredi_log(seznam, log, zacetni_n):
    for indeks, besede in seznam.iteritems():
        n = zacetni_n
        while besede > (log ** n):
            n += 1
        seznam[indeks] = log ** n
    
    return seznam
```

```python
fics_besede_log = naredi_log(fics['besede'].copy(), 2, 10)
delitelji = fics_besede_log.groupby(fics_besede_log).size()

fics_zaključene_besede_log = naredi_log(fics[fics['zaključen'] == True]['besede'].copy(), 2, 10)
imenovalci = fics_zaključene_besede_log.groupby(fics_zaključene_besede_log).size()

(imenovalci / delitelji).plot.bar()
```

Vidimo zanimiv vzorec:

Zelo malo ficov dolžin med 33.000 in 260.000 besedami je zaključenih, in bolj kot se od te sredine oddaljujemo, bolj verjetno so fici zaključeni

To bi bilo najbolje predstaviti z 3d bar plotom.

```python
graf3 = avtor_questi_fici.groupby(['questi', 'fici']).size().unstack().fillna(0)
graf3
```

```python
g3 = graf3.to_numpy()

zgornja_meja = 50

fig = plt.figure(figsize=(16, 6))
ax = fig.add_subplot(121,projection='3d')
ax2 = fig.add_subplot(122,projection='3d')

lx= len(g3[0])
ly= len(g3[:,0])
xpos = np.arange(lx)
ypos = np.arange(ly, 0, -1)
xpos, ypos = np.meshgrid(xpos, ypos)

xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros(lx*ly)

dx = 0.5 * np.ones_like(zpos)
dy = dx.copy()
dz = g3.flatten()
dz2 = dz.copy()
dz2[dz2 > zgornja_meja] = zgornja_meja


ax.bar3d(xpos,ypos,zpos, dx, dy, dz)
ax2.bar3d(xpos,ypos,zpos, dx, dy, dz2)

#sh()
#ax.w_xaxis.set_ticklabels(column_names)
ax.w_yaxis.set_ticklabels(np.arange(ly, -1, -1))
ax.set_xlabel('fici')
ax.set_ylabel('questi')
ax.set_zlabel('avtorji')
ax2.w_yaxis.set_ticklabels(np.arange(ly, -1, -1))
ax2.set_xlabel('fici')
ax2.set_ylabel('questi')
ax2.set_zlabel('avtorji')

plt.show()
```

Za boljšo preglednost sem dodal še graf, kjer višina stolpcev omejena z zgornjo mejo.

Zdi se, da večina avtorjev piše večinoma samo fice ali samo queste. Ne zdi se, kot da bi več ficov pomenilo tudi več questov, kvečjemu nasprotno.

## Fici po tednih

S podatkov lahko razberemo, kako so se posamezni fici pisali:

```python
def nariši_graf_ficov(seznam_imen_ficov, stolpec):
    '''sprejme ime seznam ficov in
    stolpec po katerem delamo graf.
    Smiselni stolpci so:
    nova poglavja,
    nove besede,
    poglavja,
    besede'''
    
    prejsnji_podatki = pd.DataFrame()
    
    for fic in seznam_imen_ficov:
        podatki_fic = po_tednih[(po_tednih['naslov'] == fic) & (po_tednih['nova poglavja'] < 20)][['teden zadnje objave', stolpec]]
        # Tukaj < 20 pogalvij, ker na nekaterih mestih fici napisani kot da so popolnoma novi, kar daje napačne rezultate
        podatki_fic['teden zadnje objave'] = pd.to_datetime(podatki_fic['teden zadnje objave'], format='%Y-%m-%d')
        if prejsnji_podatki.empty:
            prejsnji_podatki = podatki_fic.rename({stolpec: stolpec + ' ' + fic}, axis='columns')
        else:
            prejsnji_podatki = pd.merge(prejsnji_podatki, podatki_fic, left_on='teden zadnje objave', right_on='teden zadnje objave', how='outer').rename({stolpec: stolpec + ' ' + fic}, axis='columns').sort_values(by = ['teden zadnje objave'])
    
    if ((stolpec == 'besede') or (stolpec == 'poglavja')):
        # za skupno besede in poglavja smiselno zapolniti z zadnjim podatkom
        prejsnji_podatki = prejsnji_podatki.fillna(method='ffill')
    else:
        # za nove besede in poglavja smiselno napolniti z 0
        prejsnji_podatki = prejsnji_podatki.fillna(0)
    
    return prejsnji_podatki.plot('teden zadnje objave')
```

```python
po_tednih.groupby('naslov').size().sort_values(ascending = False)[0:3]
```

```python
nariši_graf_ficov(['Mauling Snarks', 'Ring-Maker', 'Abaddon Born(e)'], 'besede')
```

```python
nariši_graf_ficov(['Mauling Snarks', 'Ring-Maker', 'Abaddon Born(e)'], 'nove besede')
```

## Avtorji po tednih

Če hočemo primerjati vse fice enega avtorja:

```python
po_tednih.groupby('avtor').size().sort_values(ascending = False)[0:3]
```

```python
def nariši_graf_avtor(avtor, stolpec):
    '''sprejme ime avtorja in 
    stolpec po katerem delamo graf.
    Smiselni stolpci so:
    nova poglavja,
    nove besede,
    poglavja,
    besede'''
    
    prejsnji_podatki = pd.DataFrame()
    
    for fic, _ in po_tednih[po_tednih['avtor'] == avtor].groupby('naslov').size().iteritems():
        podatki_fic = po_tednih[(po_tednih['naslov'] == fic) & (po_tednih['nova poglavja'] < 20)][['teden zadnje objave', stolpec]]
        podatki_fic['teden zadnje objave'] = pd.to_datetime(podatki_fic['teden zadnje objave'], format='%Y-%m-%d')
        if prejsnji_podatki.empty:
            prejsnji_podatki = podatki_fic.rename({stolpec: stolpec + ' ' + fic}, axis='columns')
        else:
            prejsnji_podatki = pd.merge(prejsnji_podatki, podatki_fic, left_on='teden zadnje objave', right_on='teden zadnje objave', how='outer').rename({stolpec: stolpec + ' ' + fic}, axis='columns').sort_values(by = ['teden zadnje objave'])
        
    if ((stolpec == 'besede') or (stolpec == 'poglavja')):
        # za skupno besede in poglavja smiselno zapolniti z zadnjim podatkom
        prejsnji_podatki = prejsnji_podatki.fillna(method='ffill')
    else:
        # za nove besede in poglavja smiselno napolniti z 0
        prejsnji_podatki = prejsnji_podatki.fillna(0)
        
    return prejsnji_podatki.plot('teden zadnje objave')
```

```python
nariši_graf_avtor('The Laurent', 'besede')
```

```python
nariši_graf_avtor('The Laurent', 'nove besede')
```
