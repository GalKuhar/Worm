```python
import pandas as pd
pd.options.display.max_rows = 20
```

```python
fics = pd.read_csv('obdelani_podatki\podatki_vseh_ficov.csv', index_col='naslov' )
strani = pd.read_csv('obdelani_podatki\podatki_strani.csv', index_col='naslov' )
po_tednih = pd.read_csv('obdelani_podatki\podatki_po_tednih.csv', index_col='naslov' )
```

```python
avtorji = fics.groupby('avtor')
```

```python
a1 = avtorji['besede'].sum()
```

```python
a1[0:10]
```

```python
a2 = avtorji.size().rename('fici')
```

```python
a2[0:10]
```

```python
graf1 = pd.concat([a1, a2], axis=1)
graf1
```
